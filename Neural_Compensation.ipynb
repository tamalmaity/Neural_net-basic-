```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
from tensorflow.keras.callbacks import TensorBoard

NAME = "neural-comp-visualize-{}".format(int(time.time()))
tensorboard = TensorBoard(log_dir = 'logs/{}'.format(NAME))
```

```python
data_x = pd.read_excel(r"/media/tamal/New Volume/Windows/Downloads/vector_voltmeter.xlsx", sheet_name = "Sheet1",
                     usecols = "B,G:H")
data_x = data_x.values[0:364]
data_y = pd.read_excel(r"/media/tamal/New Volume/Windows/Downloads/vector_voltmeter.xlsx", sheet_name = "Sheet1",
                     usecols = "E:F")
data_y = data_y.values[0:364] #364 values in total
data = np.concatenate((data_x,data_y), axis = 1)

data_train = data[0:291] # 80% of data,i.e., 291 values
data_test = data[291:364] # 20% of data, i.e., 73 values 

x_train = data_train[:,0:3]
y_train = data_train[:,3:5]

x_test = data_test[:,0:3]
y_test = data_test[:,3:5]


#x_train = tf.keras.utils.normalize (x_train, axis = 1)
#x_test = tf.keras.utils.normalize (x_test, axis = 1)

#y_train = tf.keras.utils.normalize (x_train, axis = 1)
#y_test = tf.keras.utils.normalize (x_test, axis = 1)
```

```python
model = tf.keras.models.Sequential()
tf.contrib.layers.xavier_initializer(
    uniform=True,
    seed=None,
    dtype=tf.dtypes.float32
)

tf.train.exponential_decay(
    learning_rate = 0.000001,
    global_step = 1,
    decay_steps = 10000,
    decay_rate = 0.9,
    staircase=False,
    name=None
)


model.add(tf.keras.layers.Dense (10, activation = tf.nn.relu))
#model.add(tf.keras.layers.Dense (10, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense (2, activation = None))

model.compile(optimizer = 'adam',
             loss = 'mean_squared_error',
             metrics = ['accuracy'])

model.fit(x_train,y_train,epochs = 300, callbacks = [tensorboard]) #to see graphs goto the directory housing logs and do this --> tensorboard --logdir='logs/' --> then open the localhost link 
```

```python
val_loss, val_acc = model.evaluate(x_test,y_test)
print(val_loss, val_acc)
```
